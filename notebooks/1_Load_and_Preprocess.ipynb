```python

import nltk                        
# Download datasets inside NLTK 
nltk.download('reuters')          # Reuters news articles dataset
nltk.download('stopwords')        # List of common stopwords 

from nltk.corpus import reuters   
from nltk.corpus import stopwords 
from nltk.stem import PorterStemmer  # Stemming = reduce words 
import re                         # Regular expressions for text cleaning


docs = [reuters.raw(fid) for fid in reuters.fileids()]
# docs = a list that contains the FULL TEXT of each economic article

file_ids = reuters.fileids()
# file_ids = a list of IDs identifying each article
           
```

```python
#preprocessing function 
stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()

def preprocess(text):
    text = text.lower()

    #  remove punctuation & numbers (keep only letters and spaces)
    text = re.sub(r'[^a-z\s]', '', text)
    
    tokens = text.split()
    
    cleaned = [stemmer.stem(word) for word in tokens if word not in stop_words]
    
    return cleaned

```

```python
# save file ID s and cleaned docs 

import json
import pickle
cleaned_docs = []

for i, doc in enumerate(docs):
    tokens = preprocess(doc)
    cleaned_docs.append(" ".join(tokens))   

with open("file_ids.json", "w") as f:
    json.dump(file_ids, f)

with open("cleaned_docs.pkl", "wb") as f:
    pickle.dump(cleaned_docs, f)

```
