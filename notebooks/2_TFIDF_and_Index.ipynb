```python
import json
import pickle

# load file IDs
with open("file_ids.json", "r") as f:
    file_ids = json.load(f)

# load cleaned documents
with open("cleaned_docs.pkl", "rb") as f:
    cleaned_docs = pickle.load(f)
```

```python
# Inverted index = { term: {doc_id: term_frequency} }

inverted_index = {}

for doc_id, text in zip(file_ids, cleaned_docs):
    words = text.split()
    for word in words:
        if word not in inverted_index:
            inverted_index[word] = {}
        if doc_id not in inverted_index[word]:
            inverted_index[word][doc_id] = 0
        inverted_index[word][doc_id] += 1

# Save index to JSON file
with open("inverted_index.json", "w") as f:
    json.dump(inverted_index, f)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy import sparse
# Create TF-IDF vectorizer
vectorizer = TfidfVectorizer(stop_words='english', min_df=3)

# Transform documents into TF-IDF matrix
X = vectorizer.fit_transform(cleaned_docs)

#save the matrix 
sparse.save_npz("tfidf_matrix.npz", X)
```
